## 1.0 Import Libraries

```python
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics as mt
```

### 1.1 Load Dataset

```python
dataset_path = '../dataset/train.csv'

df = pd.read_csv(dataset_path)
```

```python
df.head()
```

```python
df.loc[:, 'limite_adicional'].unique()
```

```python
# verificando o balanceamento das classes
df.loc[:, 'limite_adicional'].value_counts(normalize=True)
```

## 2.0 Seleção de Features

```python
df.columns
```

```python
#x = features_treinamento
#y = label

features_treinamento = ['idade', 'saldo_atual', 'divida_atual', 'renda_anual',
                        'valor_em_investimentos', 'taxa_utilizacao_credito', 'num_emprestimos',
                        'num_contas_bancarias', 'num_cartoes_credito', 'dias_atraso_dt_venc',
                        'num_pgtos_atrasados', 'num_consultas_credito', 'taxa_juros']
label = 'limite_adicional'

x_train = df.loc[:, features_treinamento]
y_train = df.loc[:, label].values
```

```python
x_train
```

```python
y_train
```

## 3.0 Treinamento

```python
#definição de parametros do treinamento
k = 7
knn_classifier = KNeighborsClassifier(n_neighbors=k)

#treinamento do algoritmo
knn_classifier.fit(x_train, y_train)
```

```python
y_pred_7 =knn_classifier.predict(x_train)
```

```python
df_result = df.copy()
df_result['classificacao_k_7'] = y_pred_7
```

```python
df_result.loc[:, ['id_cliente','limite_adicional', 'classificacao_k_7']].sample(10)
```

## 4.0 Performance

#### 4.1 Matriz de Confusão

```python
mt.confusion_matrix(y_train, y_pred_7)
```

### 4.2 Acurácia

```python
acuracia_7 = mt.accuracy_score(y_train, y_pred_7)*100
```

```python
df_result['acertos'] = (df_result.loc[:, ['id_cliente','limite_adicional', 'classificacao_k_7']]
                        .apply(lambda x: 1 if x['limite_adicional'] == x['classificacao_k_7'] else 0, axis=1))
```

```python
df_result.loc[:, ['id_cliente', 'idade', 'limite_adicional', 'classificacao_k_7', 'acertos']].sample(10)
```

```python
print('Acuracia: {:.2f}%'.format((df_result['acertos'].sum() / df_result.shape[0])*100))
```

### 4.3 Precisão

```python
precisao = mt.precision_score(y_train, y_pred_7, pos_label='Conceder')
print('Precisão: {:.2f}%'.format(precisao*100))
```

### 4.4 Recall

```python
recall = mt.recall_score(y_train, y_pred_7, pos_label='Conceder')
print('Recall: {:.2f}%'.format(recall*100))
```

```python
def treinar_knn(k, x_train, y_train):
    """
    Função para treinar o modelo KNN
    :param k: número de vizinhos
    :param x_train: features de treinamento
    :param y_train: label de treinamento
    :return: acurácia do modelo
    """
    knn_classifier = KNeighborsClassifier(n_neighbors=k)
    knn_classifier.fit(x_train, y_train)
    y_pred = knn_classifier.predict(x_train)
    acuracia = mt.accuracy_score(y_train, y_pred)*100
    return acuracia
```

```python
#treinando o modelo com diferentes valores de k
knn_11 = treinar_knn(11, x_train, y_train)
knn_13 = treinar_knn(13, x_train, y_train)
knn_15 = treinar_knn(15, x_train, y_train)
knn_17 = treinar_knn(17, x_train, y_train)
knn_19 = treinar_knn(19, x_train, y_train)
knn_21 = treinar_knn(21, x_train, y_train)
```
